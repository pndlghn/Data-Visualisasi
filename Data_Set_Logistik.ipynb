```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path file yang diunggah
file_path = "Data Set Logistik (1).xlsx"

# Membaca file Excel
df = pd.read_excel(file_path)

# Menyimpan sebagai CSV
csv_file_path = "Data_Set_Logistik.csv"
df.to_csv(csv_file_path, index=False)

# Mengubah nama kolom agar sesuai dengan kode yang diberikan
df.rename(columns={
    'Berat Barang (kg)': 'Berat Barang',
    'Biaya Pengiriman (Rp)': 'Biaya Pengiriman',
    'Status Pengiriman': 'Status Pengiriman',
    'Lokasi Asal': 'Lokasi Asal'
}, inplace=True)
```

```python
print("Diagram Batang - Jumlah pesanan per Lokasi Asal")
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Lokasi Asal', hue='Lokasi Asal', palette='viridis', legend=False)
plt.title('Jumlah Pesanan per Lokasi Asal')
plt.xlabel('Lokasi Asal')
plt.ylabel('Jumlah Pesanan')
plt.xticks(rotation=45)
plt.show()
```

```python
print("Pie Chart - Proporsi status pengiriman")
plt.figure(figsize=(8,8))
df['Status Pengiriman'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Proporsi Status Pengiriman')
plt.ylabel('')
plt.show()
```

```python
print("Histogram - Distribusi Biaya Pengiriman")
plt.figure(figsize=(10,5))
sns.histplot(df['Biaya Pengiriman'], bins=20, kde=True, color='blue')
plt.title('Distribusi Biaya Pengiriman')
plt.xlabel('Biaya Pengiriman')
plt.ylabel('Frekuensi')
plt.show()
```

```python
print("Scatter Plot - Hubungan Berat Barang dan Biaya Pengiriman")
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='Berat Barang', y='Biaya Pengiriman', color='red', alpha=0.6)
plt.title('Hubungan Berat Barang dan Biaya Pengiriman')
plt.xlabel('Berat Barang')
plt.ylabel('Biaya Pengiriman')
plt.show()
```

```python
print("Analisis pola data")
lokasi_terbanyak = df['Lokasi Asal'].value_counts().idxmax()
status_terbanyak = df['Status Pengiriman'].value_counts().idxmax()
biaya_median = df['Biaya Pengiriman'].median()
korelasi = df[['Berat Barang', 'Biaya Pengiriman']].corr().iloc[0, 1]
pola_korelasi = "positif" if korelasi > 0 else "negatif"

# Menampilkan hasil analisis
print("\n🔍 **Analisis Pola yang Ditemukan:**")
print(f"1️⃣ Dari bar chart, lokasi asal dengan jumlah pesanan terbanyak adalah {lokasi_terbanyak}")
print(f"2️⃣ Dari pie chart, status pengiriman yang paling sering muncul adalah {status_terbanyak}")
print(f"3️⃣ Dari histogram, biaya pengiriman terbanyak berada di rentang sekitar Rp{biaya_median}")
print(f"4️⃣ Dari scatter plot, terdapat hubungan antara berat barang dan biaya pengiriman dengan pola yang cenderung {pola_korelasi}")
```
